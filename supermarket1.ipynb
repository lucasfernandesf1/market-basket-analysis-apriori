```python
from scipy.io import arff
import pandas as pd
```

```python
data = arff.loadarff('supermarket.arff')
```

```python
df = pd.DataFrame(data[0])
```

```python
df.head()
```

```python
##transformar em strings
for col in df.columns:
    df[col] = df[col].astype(str)
```

```python
## limpando os valores
for col in df.columns:
    df[col] = df[col].str.replace("b'", "")
    df[col] = df[col].str.replace("'", "")
```

```python
df.head()
```

```python
df = df.loc[:, ~df.columns.str.contains('department')]
```

Foram removidas colunas do tipo "department", pois não representam itens de compra relevantes para análise de associação.

```python
df.shape
```

```python
df = df.loc[:, (df == 't').any()]
```

Foram removidas colunas sem ocorrência de compras (sem valores 't'), pois não contribuem para a geração de regras de associação.

```python
df.shape
```

```python
df.to_csv('supermarket_clean.csv', index=False)
```
